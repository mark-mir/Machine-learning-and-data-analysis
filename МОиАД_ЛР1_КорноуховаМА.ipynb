##8. Преобразуем значения price в числа с плавающей запятой

##9. Вычислим для столбца price среднее значение, моду и медиану

```python
mean_price = df['price'].mean()   #среднее значение
mode_price = df['price'].mode()[0]   #мода
median_price = df['price'].median(0) #медиана
print(f'Среднее значение: {mean_price}, мода: {mode_price}, медиана: {median_price}')
```

##10. Определим для столбца price дисперсию и среднеквадратичное отклонение

```python
variance_price = df['price'].var()
std_dev_price = df['price'].std()
print(f'Дисперсия: {variance_price}, среднеквадратичное отклонение: {std_dev_price}')
```

##11. Как влияет высота над уровнем моря на цену жилья?

```python
corr = df['sqft_above'].corr(df['price'])
corr
```

Воспользуемся встроенным расчётом корреляции Пирсона.

Значение 0.606 означает среднюю связь, оно положительно, значит связь прямая. Есть зависимость - чем больше высота над уровнем моря, тем больше цена.
